##Analyse Training and Testing datasets

```python
path='/content/drive/MyDrive/workflow_wiz/'
url='https://raw.githubusercontent.com/heshadharani/workflow/main/data/workflow_wiz.csv'
#Load the data grid
#df= pd.read_csv(path+'workflow_wiz.csv')
df=pd.read_csv(url)
df
```

```python
df.info()
```

```python
#Process DataFrame 
i=1

features=['wf','rank','axn','axb','axi', 'axe', 'uxb', 'uxi','uxe', 'txu', 'txd', 'txm','txud', 'txum', 'txdm', 'txudm']
s='['
for index, row in df.iterrows():
    rs='{'
    for f in features:
      rs=rs+  '"'+ f +'":"' + str(row[f]) +'",'  # build record string
    
    rs = rs[:-1] # remove last comma 
    s=s+ rs + '}'+','
    i=i+1
s=s[:-1]+']'

print(s)
# save json object to disk
jf=path+'workflow_wiz.json'
with open(jf, 'w') as f:
    f.write(s)
```

```python
url2='https://raw.githubusercontent.com/heshadharani/workflow/main/data/workflow_details.csv'
#df= pd.read_csv(path+'workflow_details.csv')
df= pd.read_csv(url2)
df
```

```python
#Process DataFrame 
i=1

features=['wf','title','des','img']
s='['
for index, row in df.iterrows():
    rs='{'
    for f in features:
      rs=rs+  '"'+ f +'":"' + str(row[f]) +'",'  # build record string
    
    rs = rs[:-1] # remove last comma 
    s=s+ rs + '}'+','
    i=i+1
s=s[:-1]+']'

print(s)
# save json object to disk
jf=path+'workflow_wiz.json'
with open(jf, 'w') as f:
    f.write(s)
```
